# Data Wrangling - GoBike Project 

Data Wrangling process from GoBike Project.

As there were many issues to be fixed in each of the 12 datasets, in order to merge in a single dataset, this file had to be created. Each of the csv file had different kind of issues, which explaind why this data wrangling process had to be repeated on each file. Otherwise, a function would be constructed to automate the whole process.
